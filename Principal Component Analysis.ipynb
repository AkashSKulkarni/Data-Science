```python
# Importing all the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
# Reading the dataset
Wine = pd.read_csv('wine.csv')
Wine
```

```python
Wine.describe()
```

```python
Wine_normal = scale(Wine)
Wine_normal
```

```python
pca = PCA(n_components = 14)
pca_values = pca.fit_transform(Wine_normal)
pca_values
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1,color = 'red')
```

```python
#Using first 3 Principal Component Scores
```

```python
pca_values[:,0:3]
```

```python
final = pd.concat([pd.DataFrame(pca_values[:,0:3],columns = ['pc1','pc2','pc3']),Wine[['Type']]],axis = 1)
final
```

```python
plt.style.use('classic')
```

```python
import seaborn as sns
```

```python
import seaborn as sns
sns.scatterplot(data = final,x = 'pc1',y = 'pc2',hue = 'Type',s = 100)  
```

```python
p1 = sns.scatterplot(data = final,x = 'pc1',y = 'pc2',s = 100)  
for line in range(0,final.shape[0]):
     p1.text(final.pc1[line], final.pc2[line],final.Type[line], horizontalalignment = 'left', size = 'medium')
```

## Hierarchical Clustering using first 3 principal component scores

```python
Wine2 = final.iloc[:,0:3]
Wine2
```

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(Wine2,method='complete',metric='euclidean')
plt.figure(figsize=(15,5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete = AgglomerativeClustering(n_clusters=4,linkage='complete',affinity='euclidean').fit(Wine2)
cluster_labels = pd.Series(h_complete.labels_)
cluster_labels
Wine2['clust'] = cluster_labels
Wine2
```

```python
Wine2.iloc[:,0:3].groupby(Wine2.clust).mean()
```

## K-Means Clustering using first 3 principal component scores

```python
from sklearn.cluster import KMeans
```

```python
Wine2
```

```python
Wine3 = Wine2.iloc[:,0:3]
Wine3
```

```python
plt.figure(figsize=(15,5))
WCSS = []
for i in range(1,11):
    abc = KMeans(n_clusters=i)
    abc.fit(Wine3)
    WCSS.append(abc.inertia_)
plt.plot(range(1,11),WCSS)
plt.title('The Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
abc = KMeans(n_clusters=4)
y_KMeans = abc.fit_predict(Wine3)
y_KMeans
```

```python
md = pd.Series(y_KMeans)
Wine3['Clust'] = md
Wine3
```

```python
Wine3.iloc[:,:].groupby(Wine3.Clust).mean()
```

```python
Wine3.plot(x='pc1',y='pc2',c=abc.labels_,kind='scatter',s=50,cmap=plt.cm.coolwarm)
plt.title('Clustering using KMeans')
```
